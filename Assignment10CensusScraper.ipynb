```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import re
from time import sleep, time
from random import randint
```

```python
#FIPS id for every county in Michigan
fips = [26131,26013,26053,26095,26153,26071,26043,26109,26031,26141,26009,26137,26119,26001,26079,26039,26135,26019,26101,26165,26143,26113,26069,26129,26105,26085,26133,26035,26051,26017,26111,26127,26073,26107,26123,26157,26151,26145,26121,26117,26057,26087,26081,26049,26139,26155,26067,26037,26099,26125,26093,26065,26045,26015,26005,26161,26075,26025,26159,26077,26021,26091,26023,26059,26149,26027,26083,26061,26103,26003,26041,26033,26097,26047,26029,26089,26055,26007,26011,26063,26147,26163,26115]

print(len(fips))
```

```python
#Comma separated fips codes can be used, but only up to 6 counties at a time unforunately 
#So I think I'll just query every county independently
baseUrl = "https://www.census.gov/quickfacts/fact/table/"
```

```python
#We'll store FIPS, name, area, and population data from the url
df = pd.DataFrame(columns=["FIPS", "County name", "Population est 2019", "Area"])

DEBUG = True
for id in fips:
    sourceUrl = baseUrl + str(id)
    html = urlopen(sourceUrl)
    soup = BeautifulSoup(html)
    #I will fill this data in, then add it to the dataframe
    FIPS_INDEX = 0
    COUNTY_NAME_INDEX = 1
    POP_INDEX = 2
    AREA_INDEX = 3
    
    #The text labels for each value are for readability, the df's column labels are the important labels
    values = [("FIPS", id), ("County name", None), ("Population est 2019", None), ("Area", None)]
    
    #Get population
    #TODO this is the default selected fact? Seems to be? It'd be better if I did this a different way
    popRow = soup.find_all(class_="fact selected")[0]
    expr = "data-value=\"(\d+)\""
    popReg=re.search(expr, str(popRow))
    pop = int(popReg.group(1))
    if(DEBUG):
        print(str(popRow)+"\n")
        print("pop type=", type(pop))
        print("Found pop", pop)
        print("\n\n")
    #Create a new tuple with the label and the number, replace the old tuple
    values[POP_INDEX] = (values[POP_INDEX][0], pop)
    
    #Get area
    areaRow = soup.find_all('tr', {"data-unit":"SQM"})[0]
    print(str(areaRow)+"\n")
    expr = "data-value=\"(\d+.*\d*)\""
    areaReg=re.search(expr, str(areaRow))
    area = float(areaReg.group(1))
    if(DEBUG):
        print(str(areaRow)+"\n")
        print("area type=", type(area))
        print("Found area", area)
        print("\n\n")
    #Create a new tuple with the label and the number, replace the old tuple
    values[AREA_INDEX] = (values[AREA_INDEX][0], area)
    
    #Get name
    nameRow = soup.find_all(class_="qf-geobox")[0]
    print(str(nameRow)+"\n")
    name = nameRow.findChildren("span")[0].text
    #County name is followed by this string, it's redundant for my purposes
    removeText = ", Michigan"
    name = name[:len(name)-len(removeText)]
    if(DEBUG):
        print(str(nameRow)+"\n")
        print("name type=", type(name))
        print("Found name", name)
        print("\n\n")
    #Create a new tuple with the label and the number, replace the old tuple
    values[COUNTY_NAME_INDEX] = (values[COUNTY_NAME_INDEX][0], name)
    
    if(DEBUG):
        print("Adding to dataframe: ", values)
    df.loc[id] = [value[1] for value in values]
    #Avoid spamming them
    sleep(randint(2,5))
```
